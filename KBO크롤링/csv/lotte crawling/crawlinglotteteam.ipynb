```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
target_url = 'http://www.giantsclub.com/html/?pcode=440'
res = urlopen( target_url )
soup = BeautifulSoup( res, 'html5lib')
soup
```

```python
output = list()

for tr in soup.select('.tbl-stat > tbody > tr'):
    output.append({
        '연도': tr.select_one('td:nth-child(1)').string,
        '순위' : tr.select_one('td:nth-child(2)').string,
        '경기' : tr.select_one('td:nth-child(3)').string,
        '승리' : tr.select_one('td:nth-child(4)').string,
        '패전' : tr.select_one('td:nth-child(5)').string,
        '무' : tr.select_one('td:nth-child(6)').string,
        '승률' : tr.select_one('td:nth-child(7)').string,
        '득점' : tr.select_one('td:nth-child(8)').string,
        '실점' : tr.select_one('td:nth-child(9)').string,
        '홈런' : tr.select_one('td:nth-child(10)').string,
        '도루' : tr.select_one('td:nth-child(11)').string,
        '타율' : tr.select_one('td:nth-child(12)').string,
        '방어율' : tr.select_one('td:nth-child(13)').string
    })
    
output
```

```python
import json
with open('res.json', 'w') as f:
    json.dump(output, f)
```

```python
from selenium import webdriver as wd
import pandas as pd
```

```python
driver = wd.Chrome('C:/Users/sxxhn/Desktop/lotte/chromedriver.exe')
```

```python
import time
for n in range(2):
    target_site = 'http://www.giantsclub.com/html/?pcode=292'
    driver.get(target_site)
    time.sleep(5)
```

```python
YEARS = [ option.get_attribute('value')
 for option in driver.find_elements_by_css_selector('#year_b > option')
    if option.get_attribute('value') 
]
YEARS[:2]
```

```python
data = list()
for year in YEARS[0:6]:
    year_sel = driver.find_element_by_id('year_b')
    year_sel.send_keys( year )
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="cont_wrapper"]/div/div[6]/div/form/input').click()
    time.sleep(5)
    
    # 데이터
    
    results = list()
    for td in driver.find_elements_by_css_selector('#tbl-rank2 > tbody > tr') :
        results.append(td.text.split())
        
    for i in range(0, len(results)) :
        data.append({
            '연도' : year,
            '팀' : results[i][0],
            '경기수' : results[i][1],
            '타수' : results[i][2],
            '안타' : results[i][3],
            '2루타' : results[i][4],
            '3루타' : results[i][5],
            '홈런' : results[i][6],
            '타점' : results[i][7],
            '볼넷' : results[i][8],
            '사구' : results[i][9],
            '삼진' : results[i][10],
            '득점' : results[i][11]
        })
```

```python
df = pd.DataFrame(data)
df
```

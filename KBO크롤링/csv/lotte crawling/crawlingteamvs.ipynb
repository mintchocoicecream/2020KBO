```python
from selenium import webdriver as wd
import pandas as pd
```

```python
driver = wd.Chrome('C:/Users/sxxhn/Desktop/lotte/chromedriver.exe')
```

```python
import time
for n in range(2):
    target_site = 'http://www.giantsclub.com/html/?pcode=292'
    driver.get(target_site)
    time.sleep(5)
```

```python
YEARS = [ option.get_attribute('value')
 for option in driver.find_elements_by_css_selector('#year_p > option')
    if option.get_attribute('value') 
]
YEARS[:2]
```

```python
data = list()
for year in YEARS[0:6]:
    year_sel = driver.find_element_by_id('year_p')
    year_sel.send_keys( year )
    driver.find_element_by_class_name('vct').click()
    time.sleep(5)
    
    # 데이터
    
    results = list()
    for td in driver.find_elements_by_css_selector('#tbl-rank > tbody > tr') :
        results.append(td.text.split())
        
    for i in range(0, len(results)) :
        if len(results[i]) == 18 :
            data.append({
                '연도' : year,
                '팀' : results[i][0],
                '경기수' : results[i][1],
                '타수' : results[i][2],
                '피안타' : results[i][3],
                '자책점' : results[i][4],
                '승' : results[i][5],
                '패' : results[i][6],
                '세이브' : results[i][7],
                '홀드' : results[i][8],
                '이닝' : (results[i][9]+" "+results[i][10]),
                '탈삼진' : results[i][11],
                '볼넷' : results[i][12],
                '방어율' : results[i][13],
                '피홈런' : results[i][14],
                '사구' : results[i][15],
                '실점' : results[i][16],
                '피안타율' : results[i][17]
            })
        else :
            data.append({
                '연도' : year,
                '팀' : results[i][0],
                '경기수' : results[i][1],
                '타수' : results[i][2],
                '피안타' : results[i][3],
                '자책점' : results[i][4],
                '승' : results[i][5],
                '패' : results[i][6],
                '세이브' : results[i][7],
                '홀드' : results[i][8],
                '이닝' : results[i][9],
                '탈삼진' : results[i][10],
                '볼넷' : results[i][11],
                '방어율' : results[i][12],
                '피홈런' : results[i][13],
                '사구' : results[i][14],
                '실점' : results[i][15],
                '피안타율' : results[i][16]
            })
```

```python
df = pd.DataFrame(data)
df
```

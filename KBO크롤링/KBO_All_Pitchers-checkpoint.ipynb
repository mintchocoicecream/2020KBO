```python
from selenium import webdriver as wd
import pandas as pd
import time
```

```python
driver = wd.Chrome('C:/Users/sxxhn/Desktop/lotte/chromedriver.exe')
```

```python
for n in range(2):
    target_site = 'https://www.koreabaseball.com/Record/Player/PitcherBasic/Basic1.aspx'
    driver.get(target_site)
    time.sleep(5)
```

```python
YEARS = [ option.get_attribute('value')
 for option in driver.find_elements_by_css_selector('#cphContents_cphContents_cphContents_ddlSeason_ddlSeason > option')
    if option.get_attribute('value') 
]
YEARS[-7:-1]
```

```python
data = list()
for year in YEARS[-7:-1]:
    # 연도 선택 
    #cphContents_cphContents_cphContents_ddlSeason_ddlSeason > option:nth-child(34) = 2015
    year_sel = driver.find_element_by_id('cphContents_cphContents_cphContents_ddlSeason_ddlSeason')
    # 2. 요소의 값을 변경한다(입력한다)
    year_sel.send_keys( year )
    time.sleep(3)
    
    
    #팀 선택
    
    for j in range( 2, 12 ):
        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ddlTeam_ddlTeam > option:nth-child('+str(j)+')').click()
        time.sleep(3)

        results = list()
        for i in range(1,30):
            for td in driver.find_elements_by_css_selector('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr:nth-child('+str(i)+')'):
                results.append(td.text.split())        
            
      
        for cnt in range(0, len(results)) :
            if len(results[cnt]) == 20:              
                data.append({
                    '연도' : year,
                    '선수이름' : results[cnt][1],
                    '팀명' : results[cnt][2],
                    '평균자책점' : results[cnt][3],
                    '경기' : results[cnt][4],
                    '승' : results[cnt][5],
                    '패' : results[cnt][6],
                    '세이브' : results[cnt][7],
                    '홀드' : results[cnt][8],
                    '승률' : results[cnt][9],
                    '이닝' : (results[cnt][10]+" "+results[cnt][11]),
                    '피안타' : results[cnt][12],
                    '홈런' : results[cnt][13],
                    '볼넷' : results[cnt][14],
                    '사구' : results[cnt][15],
                    '삼진' : results[cnt][16],
                    '실점' : results[cnt][17],
                    '자책점' : results[cnt][18],
                    '이닝당 출루허용률' : results[cnt][19]
                })
            else:
                data.append({
                    '연도' : year,
                    '선수이름' : results[cnt][1],
                    '팀명' : results[cnt][2],
                    '평균자책점' : results[cnt][3],
                    '경기' : results[cnt][4],
                    '승' : results[cnt][5],
                    '패' : results[cnt][6],
                    '세이브' : results[cnt][7],
                    '홀드' : results[cnt][8],
                    '승률' : results[cnt][9],
                    '이닝' : results[cnt][10],
                    '피안타' : results[cnt][11],
                    '홈런' : results[cnt][12],
                    '볼넷' : results[cnt][13],
                    '사구' : results[cnt][14],
                    '삼진' : results[cnt][15],
                    '실점' : results[cnt][16],
                    '자책점' : results[cnt][17],
                    '이닝당 출루허용률' : results[cnt][18]
                })
```

```python
from selenium import webdriver as wd
import pandas as pd
import time
```

```python
driver = wd.Chrome('C:/Users/sxxhn/Desktop/lotte/chromedriver.exe')
```

```python
for n in range(2):
    target_site = 'https://www.koreabaseball.com/Record/Player/Defense/Basic.aspx'
    driver.get(target_site)
    time.sleep(5)
```

```python
YEARS = [ option.get_attribute('value')
 for option in driver.find_elements_by_css_selector('#cphContents_cphContents_cphContents_ddlSeason_ddlSeason > option')
    if option.get_attribute('value') 
]
YEARS[-6:]
```

```python
data = list()
for year in YEARS[-6:]:
    # 연도 선택 
    year_sel = driver.find_element_by_id('cphContents_cphContents_cphContents_ddlSeason_ddlSeason')
    year_sel.send_keys( year )
    time.sleep(3)
    
    
    #팀 선택
    for j in range( 2, 12 ):
        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ddlTeam_ddlTeam > option:nth-child('+str(j)+')').click()
        time.sleep(3)

        results = list()
        
        # 1페이지 이동
        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo1').click()
        time.sleep(3)
        
        for i in range(1,31):
            for td in driver.find_elements_by_css_selector('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr:nth-child('+str(i)+')'):
                results.append(td.text.split())
                
        # 2페이지 이동
        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo2').click()
        time.sleep(3)

        for i in range(1, 31 ):
            for td in driver.find_elements_by_css_selector('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr:nth-child('+str(i)+')'):
                results.append(td.text.split())

        # 3페이지 이동
        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo3').click()
        time.sleep(3)

        for j in range( 1, 31 ):
            for td in driver.find_elements_by_css_selector('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr:nth-child('+str(i)+')'):
                results.append(td.text.split())

                    
        for cnt in range(0, len(results)) :
            if len(results[cnt]) == 18:
                data.append({
                    '연도' : year,
                    '선수이름' : results[cnt][1],
                    '팀명' : results[cnt][2],
                    '포지션' : results[cnt][3],
                    '경기' : results[cnt][4],
                    '선발경기' : results[cnt][5],
                    '이닝' : (results[cnt][6]+" "+results[cnt][7]),
                    '실책' : results[cnt][8],
                    '견제사' : results[cnt][9],
                    '자살' : results[cnt][10],
                    '보살' : results[cnt][11],
                    '병살' : results[cnt][12],
                    '수비율' : results[cnt][13],
                    '포일' : results[cnt][14],
                    '도루허용' : results[cnt][15],
                    '도루저지' : results[cnt][16],
                    '도루저지율' : results[cnt][17]
                })
            else:
                data.append({
                    '연도' : year,
                    '선수이름' : results[cnt][1],
                    '팀명' : results[cnt][2],
                    '포지션' : results[cnt][3],
                    '경기' : results[cnt][4],
                    '선발경기' : results[cnt][5],
                    '이닝' : results[cnt][6],
                    '실책' : results[cnt][7],
                    '견제사' : results[cnt][8],
                    '자살' : results[cnt][9],
                    '보살' : results[cnt][10],
                    '병살' : results[cnt][11],
                    '수비율' : results[cnt][12],
                    '포일' : results[cnt][13],
                    '도루허용' : results[cnt][14],
                    '도루저지' : results[cnt][15],
                    '도루저지율' : results[cnt][16]
                })

```

```python
# 데이터 하위값 확인
data[-10:]
```

```python
# 데이터프레임 생성
df = pd.DataFrame(data)
df
```

```python
df.to_csv('KBO_ALL_DEFENSE.csv', encoding='cp949')
```
